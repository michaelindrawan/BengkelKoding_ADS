```python
#Import library
import pandas as pd
import re
import numpy
import itertools
```

```python
#direktori data
dir_ = 'hungarian.data'
```

```python
#membaca data
with open(dir_, encoding="Latin1") as file:
    lines = [line.strip() for line in file]
lines[0:10]
```

```python
#memasukan data yang dibutuhkan
data = itertools.takewhile(
    lambda x: len(x) == 76,
    (' '.join(lines[i:(i+10)]).split() for i in range(0, len(lines), 10))
)
df = pd.DataFrame.from_records(data)
df.head()
```

```python
#menampilkan info dataFrame
df.info()
```

```python
#menghilangkan kolom yang tidak dibutuhkan
df = df.iloc[:,:-1]
df = df.drop(df.columns[0], axis = 1) 
```

```python
#mengubah tipe data ke float
df = df.astype(float)
```

```python
#menampilkan informasi dataFrame
df.info()
```

```python
# Validasi Data
```

```python
#mengganti nilai -9 dengan NaN
import numpy as np
df.replace(-9.0, np.nan, inplace=True)
```

```python
#menampilkan jumlah data yang hilang
df.isnull().sum()
```

```python
#menampilkan data
df.head()
```

```python
#menampilkan informasi dataFrame
df.info()
```

```python
#Menentukan Objek Data
```

```python
#memilih kolom yang dibutuhkan
df_selected = df.iloc[:,[1,2,7,8,10,14,17,30,36,38,39,42,49,56]]
```

```python
#menampilkan data
df_selected.head()
```

```python
#menampilkan informasi data
df_selected.info()
```

```python
#memapping nama kolom dan menggantinya
column_mapping = {
  2: 'age',
  3: 'sex',
  8: 'cp',
  9: 'trestbps',
  11: 'chol',
  15: 'fbs',
  18: 'restecg',
  31: 'thalach',
  37: 'exang',
  39: 'oldpeak',
  40: 'slope',
  43: 'ca',
  50: 'thal',
  57: 'target'
}
df_selected.rename(columns = column_mapping, inplace=True)
```

```python
#menampilkan data
df_selected.info()
```

```python
#Membersihkan Data
```

```python
#melihat jumlah nilai yang hilang
df_selected.isnull().sum()
```

```python
#menghapus baris yang memiliki nilai NaN
column_to_drop = ['ca', 'slope', 'thal']
df_selected = df_selected.drop(column_to_drop, axis=1)
```

```python
#menampilkan data yang sudah dibersihkan
df_selected.isnull().sum()
```

```python
#mengganti nilai NaN dengan mean
meanTBPS = df_selected['trestbps'].dropna()
meanChol = df_selected['chol'].dropna()
meanfbs = df_selected['fbs'].dropna()
meanRestCG = df_selected['restecg'].dropna()
meanthalach = df_selected['thalach'].dropna()
meanexang = df_selected['exang'].dropna()

meanTBPS = meanTBPS.astype(float)
meanChol = meanChol.astype(float)
meanfbs = meanfbs.astype(float)
meanthalach = meanthalach.astype(float)
meanexang = meanexang.astype(float)
meanRestCG = meanRestCG.astype(float)

meanTBPS = round(meanTBPS.mean())
meanChol = round(meanChol.mean())
meanfbs = round(meanfbs.mean())
meanthalach = round(meanthalach.mean())
meanexang = round(meanexang.mean())
meanRestCG = round(meanRestCG.mean())
```

```python
#mengganti nilai NaN dengan mean
fill_values = {
  'trestbps': meanTBPS,
  'chol': meanChol,
  'fbs': meanfbs,
  'thalach':meanthalach,
  'exang':meanexang,
  'restecg':meanRestCG
}

df_clean = df_selected.fillna(value=fill_values)
```

```python
#menampilkan data yang sudah dibersihkan
df_clean.info()
```

```python
#menampilkan data yang sudah dibersihkan
df_clean.isnull().sum()
```

```python
#menampilkan duplicate rows
duplicate_rows = df_clean.duplicated()
df_clean[duplicate_rows]
```

```python
#menghapus duplicate rows
df_clean = df_clean.drop_duplicates()
df_clean[df_clean.duplicated(keep=False)]
```
